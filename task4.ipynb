```python
#  if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv
Exceeded_expectations = df[df['CodeKata Score'] > 15000]
Exceeded_expectations.head()
```

```python
# if 10000<codekata score<15000 (Reached_expectations.csv)
Reached_expectations = df[df['CodeKata Score'].between(10000,15000)]
Reached_expectations.head()
```

```python
# if 7000<codekata score<10000 (Needs_Improvement.csv)
Needs_Improvement = df[df['CodeKata Score'].between(7000,10000)]
Needs_Improvement.head()
```

```python
# if codekate score < 7000 (Unsatisfactory.csv)
Unsatisfactory = df[df['CodeKata Score']<7000]
Unsatisfactory.head()
```

```python
# Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)
Avg_Previous_Geekions = df['Previous Geekions'].mean()
Avg_CodeKata_Score = df['CodeKata Score'].mean()
print('Avg_Previous_Geekions : ',Avg_Previous_Geekions)
print('Avg_CodeKata_Score : ',Avg_CodeKata_Score)
```

```python
df_new = df.fillna(0)
```

```python
#Average completion of python course or my_sql or python english or computational thinking
# those who r attempt 100 in course assume to be complete course proprely(ie = py_course,my_sql,py_english)
completion_py_course = df_new[(df_new['python'] == 100)]
Avg_comp_py_course = completion_py_course['python'].mean()

completion_my_sql_course = df_new[(df_new['mysql'] == 100)]
Avg_comp_mysql_course =  completion_my_sql_course['mysql'].mean()

completion_pyen_course = df_new[(df_new['python_en'] == 100)]
Avg_comp_pyen_course = completion_my_sql_course['python_en'].mean()

# for computational_thinking those who r get greater than 5, assume to be complete the course 
completion_computational_thinking_course = df_new[(df_new['computational_thinking']>5)]
Avg_comp_cm_th_course = completion_computational_thinking_course['computational_thinking'].mean() 


```

```python
#Average completion
print('Avg_comp_py_course: ',Avg_comp_py_course)
print('Avg_comp_mysql_course: ',Avg_comp_mysql_course)
print('Avg_comp_pyen_course: ',Avg_comp_pyen_course)
print('Avg_comp_cm_th_course: ',Avg_comp_cm_th_course)



```
